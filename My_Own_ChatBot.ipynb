```python
import pandas as pd
import re
from sentence_transformers import SentenceTransformer, util

# Load CSV data
df = pd.read_csv('data.csv')
data_dict = dict(zip(df['section'], df['content']))

# Load a pre-trained semantic search model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
section_embeddings = model.encode(list(data_dict.keys()))

# Function to normalize the text
def normalize_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Function to perform semantic search
def semantic_search(user_question):
    question_embedding = model.encode(user_question)

    # Find the most similar section
    similarities = util.cos_sim(question_embedding, section_embeddings)
    best_match_idx = similarities.argmax()

    # Return the best matching section's content
    return data_dict[list(data_dict.keys())[best_match_idx]]

# Chat function to ask questions and get answers from the model
def chat_with_model():
    print("Hello! You can ask me questions about myself.")
    while True:
        user_question = input("You: ")
        if user_question.lower() == 'exit':
            print("Goodbye!")
            break

        response = semantic_search(user_question)
        print(f"Bot: {response}")
```

```python
# Start the interaction
chat_with_model()
```
